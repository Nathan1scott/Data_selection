## Load the data

Load the `MentalHealthSurvey.csv` file into a pandas DataFrame.


import the pandas library, read the CSV file into a DataFrame, and then display the first 5 rows to verify the data has been loaded correctly.

```python
import pandas as pd

df = pd.read_csv('/content/MentalHealthSurvey.csv')
display(df.head())
```

## Explore the data


Display the first few rows, check the data types, and look for missing values.

```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```


Check the data types of all columns to identify any columns that might require transformation and examine the unique values for object type columns to identify inconsistencies.

```python
display(df.dtypes)
for col in df.select_dtypes(include='object').columns:
    display(f"Unique values in '{col}':")
    display(df[col].unique())
```


Based on the unique values, the 'cgpa' column needs to be transformed into a numerical format for analysis. The 'stress_relief_activities' column contains multiple categories within each entry and should be handled as a multi-label categorical feature. Other object type columns appear to be clean categorical variables. There are no missing values and no obvious outliers to handle at this stage.

```python
cgpa_mapping = {
    '0.0-0.0': 0.0,
    '1.5-2.0': 1.75,
    '2.0-2.5': 2.25,
    '2.5-3.0': 2.75,
    '3.0-3.5': 3.25,
    '3.5-4.0': 3.75
}
df['cgpa'] = df['cgpa'].replace(cgpa_mapping)
df['cgpa'] = df['cgpa'].astype(float)

# For 'stress_relief_activities', we can split the string into a list of activities
df['stress_relief_activities'] = df['stress_relief_activities'].str.split(', ')

display(df[['cgpa', 'stress_relief_activities']].head())
display(df.dtypes)
```

## Perform exploratory data analysis (EDA)


Analyze the distribution of key variables, identify relationships between variables, and create visualizations to understand the data better.


**Numerical Variables**

I need to analyze the distribution of key numerical variables and visualize them using histograms. I will select 'age', 'cgpa', 'depression', 'anxiety', 'isolation', and 'future_insecurity'.

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'cgpa', 'depression', 'anxiety', 'isolation', 'future_insecurity']
df[numerical_cols].hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```


**Categorical Variables**

need to analyze the distribution of key categorical variables and visualize them using bar plots or count plots. I will select 'gender', 'university', 'degree_level', and 'degree_major'.

```python
categorical_cols = ['gender', 'university', 'degree_level', 'degree_major']

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_cols):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

**Correlation Matrix**:

I need to analyze the relationships between numerical variables by generating a correlation matrix and visualizing it as a heatmap.

```python
plt.figure(figsize=(10, 8))
corr_matrix = df[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

***Relationship between key categorical variables and relevant numerical variables***:

I need to explore the relationship between key categorical variables and relevant numerical variables using visualizations like box plots or violin plots. I will examine the relationship between 'degree_level' and 'cgpa', and 'gender' and 'depression'.

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='degree_level', y='cgpa', palette='viridis')
plt.title('CGPA by Degree Level')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='gender', y='depression', palette='viridis')
plt.title('Depression Score by Gender')

plt.tight_layout()
plt.show()
```


I need to analyze the 'stress_relief_activities' column by counting the occurrences of each activity and visualizing the most frequent ones.

```python
from collections import Counter

all_activities = [activity for sublist in df['stress_relief_activities'] for activity in sublist]
activity_counts = Counter(all_activities)
most_common_activities = activity_counts.most_common(10) # Get top 10

activities_df = pd.DataFrame(most_common_activities, columns=['Activity', 'Count'])

plt.figure(figsize=(10, 6))
sns.barplot(data=activities_df, x='Count', y='Activity', palette='viridis')
plt.title('Most Frequent Stress Relief Activities')
plt.xlabel('Count')
plt.ylabel('Activity')
plt.show()
```
